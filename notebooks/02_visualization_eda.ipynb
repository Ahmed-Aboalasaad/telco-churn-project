## 7. Key Insights

```python
print("📊 KEY INSIGHTS FROM EDA")
print("=" * 60)
print()
print("1. CHURN CHARACTERISTICS:")
print(f"   - Overall churn rate: {df['Churn'].sum()/len(df)*100:.2f}%")
print()
print("2. HIGHEST RISK FACTORS:")
contract_churn = df[df['Contract']=='Month-to-month']['Churn'].sum() / len(df[df['Contract']=='Month-to-month']) * 100
print(f"   - Month-to-month contracts: {contract_churn:.2f}% churn")
fiber_churn = df[df['InternetService']=='Fiber optic']['Churn'].sum() / len(df[df['InternetService']=='Fiber optic']) * 100
print(f"   - Fiber optic internet: {fiber_churn:.2f}% churn")
check_churn = df[df['PaymentMethod']=='Electronic check']['Churn'].sum() / len(df[df['PaymentMethod']=='Electronic check']) * 100
print(f"   - Electronic check payment: {check_churn:.2f}% churn")
print()
print("3. LOWEST RISK FACTORS:")
two_year_churn = df[df['Contract']=='Two year']['Churn'].sum() / len(df[df['Contract']=='Two year']) * 100
print(f"   - Two-year contracts: {two_year_churn:.2f}% churn")
dsl_churn = df[df['InternetService']=='DSL']['Churn'].sum() / len(df[df['InternetService']=='DSL']) * 100
print(f"   - DSL internet: {dsl_churn:.2f}% churn")
print()
print("4. NUMERIC PATTERNS:")
avg_tenure_churn = df[df['Churn']==1]['tenure'].mean()
avg_tenure_no_churn = df[df['Churn']==0]['tenure'].mean()
print(f"   - Churned customers avg tenure: {avg_tenure_churn:.1f} months")
print(f"   - Retained customers avg tenure: {avg_tenure_no_churn:.1f} months")
print()
print("5. RECOMMENDATIONS:")
print("   ✓ Target month-to-month customers for contract conversion offers")
print("   ✓ Improve Fiber optic service quality and support")
print("   ✓ Promote automatic payment methods")
print("   ✓ Focus retention efforts on new customers (low tenure)")
```
